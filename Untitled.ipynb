```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.rename(columns={'sepal length (cm)':'sep_len',
                          'sepal width (cm)':'sep_wid',
                          'petal length (cm)':'pet_len', 
                          'petal width (cm)':'pet_wid'},
                 inplace=True)
```

```python
df['pet_area'] = df['pet_wid'] * df['pet_len']
df['sep_area'] = df['sep_wid'] * df['sep_len']
```
